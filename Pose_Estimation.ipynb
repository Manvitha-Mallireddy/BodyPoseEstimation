```python
import cv2
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
mp_drawing=mp.solutions.drawing_utils
mp_drawing_styles=mp.solutions.drawing_styles
mp_pose=mp.solutions.pose
pose_model = mp_pose.Pose(static_image_mode=True)
# Load the image
image_target = cv2.imread('target_image.png')
# Process the image
image_target.setflags(write=0)
image_target=cv2.cvtColor(image_target, cv2.COLOR_BGR2RGB)
results = pose_model.process(image_target)
# Extract the pose landmarks
image_target.setflags(write=1)
image_target=cv2.cvtColor(image_target, cv2.COLOR_RGB2BGR)
mp_drawing.draw_landmarks(
    image_target,
    results.pose_landmarks,
    mp_pose.POSE_CONNECTIONS,
    landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style()
    )
# Convert the pose landmarks to an image
#pose_image = cv2.cvtColor(pose_landmarks, cv2.COLOR_BGR2RGB)
# Display the image
#cv2.imshow('Image', image)
#cv2.imwrite('generated_pose.png',cv2.cvtColor(np.squeeze(image_target)*255,cv2.COLOR_BGR2RGB))
plt.imshow(image_target)
plt.title('Pose Estimation')
plt.show()
```
